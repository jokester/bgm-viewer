```python
import duckdb
import os
from pathlib import Path

Path.cwd()
```

```python
subjects_df = duckdb.sql("""
           SELECT *
            FROM read_json_auto('../dataset/subject.jsonlines') AS subjects
            -- GROUP BY date
            LIMIT 200
           """).df()
subjects_df
```

```python
episodes_df = duckdb.sql("""
           SELECT airdate, COUNT(1) AS cnt
            FROM read_json_auto('../dataset/episode.jsonlines') AS subjects
            GROUP BY airdate
            HAVING cnt < 100
            ORDER BY cnt DESC
           """).df()
episodes_df
```

```python
edges_df = duckdb.sql("""
           SELECT s2c.*, s.*, c.*
            FROM read_json_auto('../dataset/subject-characters.jsonlines') AS s2c
            INNER JOIN read_json_auto('../dataset/subject.jsonlines') AS s
            ON s2c.subject_id = s.id
            INNER JOIN read_json_auto('../dataset/character.jsonlines') AS c
            ON s2c.character_id = c.id
            WHERE s2c.type =4
            LIMIT 10
           """).df()
edges_df
```

```python
edges_df = duckdb.sql("""
           SELECT s2c.type, s.name AS subject_name, c.name AS character_name, s.*
            FROM read_json_auto('../dataset/subject-characters.jsonlines') AS s2c
            INNER JOIN read_json_auto('../dataset/subject.jsonlines') AS s
            ON s2c.subject_id = s.id
            INNER JOIN read_json_auto('../dataset/character.jsonlines') AS c
            ON s2c.character_id = c.id
            WHERE s2c.type = 3
            LIMIT 10
           """).df()
edges_df
```
